# Optimization of IAS w/ RAMP-wakeup signal chain
There are two key metrics in this signal chain: false rejection of true glass break events and average power consumption. 
These metrics are interrelated and the RAMP has several configuration options that play into these metrics.
These include binary options—such as whether or not to use preroll and whether or not to use multiple (i.e. fast/slow) detectors—as
well as continuous options—such as the amount of preroll compression and the latency/FAR/FRR tradeoffs of each detector.
This notebook analyzes these tradeoffs to present a recommendation for the signal chain. 

This is intended to be a living/editable document and any of the assumptions/models can be updated.

# Justification for pressure latency assumptions
Aspinity's modeling has little insight into the pressure data requirements. We have derived assumptions below
about how the latency of triggering the pressure sensor (PS) to switch to a higher sampling rate will impact the overall accuracy 
of the signal chain. 

The quick summary (full details in the remainder of this section) is that after reviewing Infineon's slides on audio/pressure 
alignment we have observed that the pressure peaks always occur at least 80ms after the start of audio (as opposed to the audio 
peak). We have assumed that at least two higher-sample-rate samples are required prior to the pressure peak; we know that 
there is a 6ms delay to switch the sample rate. Based on these assumptions/information we are assuming that if the RAMP trigger 
to switch the PS sample rate occurs within 42.75ms from the start of the event then there will be no degradation in the system 
accuracy, but if the RAMP trigger comes after that then there will be 100% degradation in the accuracy.

## Pressure requirements background
An audio-based detection from the RAMP is used to determine when to switch the pressure sensor to a higher sampling rate. 
* Pressure sensor sampling-rate modes:
  * Always-on at lower sampling rate (32sps, 31.25ms sampling period)
  * Higher-rate when detection happens (64sps, 15.625ms sampling period)
* The time for the PS to switch from always-on to higher-rate is about 6ms (@ 64sps/4OSR)

## Pressure/audio timeline
One example of the relationship between audio and pressure as captured by Infineon for a rubber ball impact is shown below. We have added circles to show an 
estimated PS sampling grid and a suggested location where the PS sampling speed might be stepped up based on a detection.



## Pressure/audio timing definitions
We have taken the example above and zoomed in to define the points of interest.

* Glass break audio starts at $t=0$. This is the point where the audio starts to diverge from its baseline.
* Audio peak happens at $t=\text{Pk}_{audio}$. The audio peak timing is heavily dependent on the type of event. 
  The plots shown are based on a rubber ball.
* The pressure sensor switches to higher sampling rate at $t=\text{PS}_{trig}=L_\text{press}+6\text{ms}$. $L_{press}$ 
  is the time when the RAMP triggers the system to switch sampling rates and $6\text{ms}$ is the time for the PS to switch from 32sps to 64sps.
* The pressure sensor needs to switch to a higher sampling rate by $t=\text{PS}_{trig,max}$ in order to capture
  sufficient pressure data to classify the event. Below we will estimate this time for use in this analysis.
* Pressure peak happens at $t=\text{Pk}_{press}>\text{Pk}_{audio}+43\text{ms}$
  * IFX tests show $t=\text{Pk}_{press}-\text{Pk}_{audio}$ ranges from 43ms to 83ms
  * After reviewing the Infineon slides, we have observed that $\text{Pk}_{press}$ ranges from 80ms to 100ms.



## Estimating $\text{PS}_{trig,max}$
Let us assume that the algorithm needs two higher-SPS samples prior to the peak (__this is open to be revised__). 
This means that $\text{PS}_{trig,max}$ will be:

```octave
% Assumptions
samples_before_peak = 2;     % number of high-sps samples that are needed prior to the pressure peak
sps_high            = 64;    % PS switches to 64sps
Pk_press_min        = 80e-3; % Pressure peak occurs at least 80ms after start of event
sps_switch_time     = 6e-3;  % 6ms to switch from 32sps to 64sps

% Calculate PS_trig,max
sample_period_high = 1/sps_high;
PS_trig_max        = Pk_press_min - (samples_before_peak * sample_period_high);
fprintf('PS_trig,max = %dms\n',PS_trig_max*1e3);

% Calculate max acceptable pressure trigger latency from RAMP
L_press_max = PS_trig_max - sps_switch_time;
fprintf('L_press,max = %dms\n',L_press_max*1e3);
```

# IAS w/ RAMP design tradeoffs
## Overview
### Output metrics
* Total system power ($P_{tot}$): Impacted by RAMP's preroll compression level and detector false alarm rate.
* System rejection rate ($\Delta\text{FR}_{tot}$): Impacted by RAMP's rejection rate, as well as IAS' rejection rate as a function of preroll/latency. Specify as rejection caused by adding us [$\Delta\text{FR}_{tot} = \text{FR}_\text{tot} - \text{FR}_\text{IAS}\left(\text{no RAMP}\right)$].

### Input variables
* Inputs to $P_{tot}$: $f_{rollAvg}$, $f_{press}$, $f_{audio}$, $f_{comp}$
* Inputs to $\Delta\text{FR}_{tot}$: $L_{comp}$, $f_{comp}$, $L_{press}$, $f_{press}$, $L_{audio}$, $f_{audio}$, $f_{rollAvg}$

## System Detection Rate ($\Delta\text{FR}_{tot}$)
For the system to detect a glass break,
* It must be detected by the RAMP detector,
* and the pressure/audio input to the IAS algo must have
  * Sufficiently low latency,
  * and/or sufficient preroll resolution.

```octave
FR_RAMP = @(L_RAMP, f_RAMP).01/(L_RAMP/.01*f_RAMP*60);
%FR_RAMP(.01,1/60)

% Load characterization model of the detector
RAMP_GB_Model_Path=expandVarPath('%AspBox%\engr\sw\proj\RAMPdev\RAMP_Operator_Examples\glass_break_examples\mharish\');
RAMP_GB_Model_File='Rampdev_results_strcuture_tes_rev_newtrain.mat';
%RAMP_GB_Model_File='Rampdev_results_strcuture_tes_rev.mat';
load([RAMP_GB_Model_Path '/' RAMP_GB_Model_File]);

FR_RAMP = @(L_RAMP, f_RAMP)FR_RAMP_Model(L_RAMP,f_RAMP,res_struct);
```

* $\Delta\text{FR}_\text{IAS}$: Additional IAS false rejection caused by detection latency and preroll.
  * Latency for the pressure sensor ($L_{press}$) (see the pressure sensor justification above), 
  * Latency for the audio chain ($L_{audio}$), and
  * Average frequency of preroll (i.e. compression), ($f_{rollAvg}$),

```octave
% Pressure latency
DelFR_IAS_press = @(L_press) 1*(L_press>PS_trig_max); % Assume it rejects everything if pressure latency exceeds PS_trig_max (defined above)
```

```octave
IasRollChar_Path=expandVarPath('%AspBox%/engr/sig_proc/Projects/External/Infineon/IAS_glassbreak/bdr/IAS_characterization');
IasRollChar_File='ias_char16k_20200630_parsed.mat';
load([IasRollChar_Path '/' IasRollChar_File]);

DelFR_IAS_audio = @(L_audio,f_rollAvg)IAS_FRR(res16k,L_audio,f_rollAvg,'old');
DelFR_IAS_audio(0.05,0)
```

```octave
% Total
DelFR_IAS = @(L_press, L_audio, f_rollAvg)min(1, DelFR_IAS_press(L_press) + DelFR_IAS_audio(L_audio,f_rollAvg));
```

* $\Delta\text{FR}_{tot}$: Additional rejection in the system caused by RAMP

```octave
DelFR_tot = @(L_RAMP, f_RAMP, L_press, L_audio, f_rollAvg) min(1,FR_RAMP(L_RAMP, f_RAMP) + DelFR_IAS(L_press, L_audio, f_rollAvg));
```

## System power ($P_{tot}$)
Definitions:

* $V_{dd}$: System supply voltage

```octave
Vdd = 1.8;
```

* $T_{fast}$: Fast timeout. If peripheral woken by fast detector, then timeout and shutoff after this time.

```octave
T_fast = 0.2;
```

* $T_{slow}$: Slow timeout. If peripheral or system work by accurate detector, then timeout and shutoff after this time.

```octave
T_slow = 1;
```

* $P_{rollAvg}$: Average power due to preroll. Includes activation of microcontroller to acquire each sample.

```octave
f_rollAvg = 100;               % 100Hz; Average frequency for acquiring preroll samples
E_rollAcq = Vdd * 15e-6 / 100; % Vdd * 15uA / 100Hz; Energy to acquire each preroll sample. Based upon STM32L4
P_rollAvg = @(f_rollAvg) f_rollAvg * E_rollAcq; 
```

* $P_{pressAvg}$: Average power of the pressure sensor (DPS310) as activated by the RAMP detector.

```octave
P_pressLow = Vdd * 121.6e-6; % Power at SR 32sps, 4OSR; 3V * 3.8uA * 32sps
P_pressHig = Vdd * 243.2e-6; % Power at SR 64sps, 4OSR; 3V * 3.8uA * 64sps

f_press = 1e-3;   % Hz; How often the pressure sensor is activated by the RAMP detection trigger
T_press = T_fast; % seconds; How long the pressure sensor is on each time it is activated

P_pressAvg = @(f_press, T_press) P_pressHig * f_press*T_press ...  % power when active
                               + P_pressLow * (1-f_press*T_press); % power when not active
```

* $P_{audioAvg}$: Average power of the mic (IM73A135V01), ADC, and audio buffering as activated by the RAMP detector.
  * For ADC and audio buffering, using 2.1mA @ 48kHz = 700uA @ 16kHz
    * https://www.st.com/resource/en/application_note/dm00367673-analogtodigital-audio-conversion-example-using-stm32l4-series-microcontroller-peripherals-stmicroelectronics.pdf

```octave
P_audioLow = Vdd * 70e-6; % Just the mic in low-power mode (70uA)
P_audioHig = Vdd * (170e-6 + 700e-6); % Mic in performance mode (170uA), ADC+buffering (assuming ~700uA)

f_audio = 1e-3;   % Hz; How often the audio chain is activated by the RAMP detection trigger
T_audio = T_fast; % seconds; How long the audio chain is on each time it is activated

P_audioAvg = @(f_audio, T_audio) P_audioHig * f_audio*T_audio ...  % power when active
                               + P_audioLow * (1-f_audio*T_audio); % power when not active
```

* $P_{compAvg}$: Average power of the compute core w/ glass break processing as activated by the RAMP detector.
  * Numbers currently used for Quicklogic, page 124 of https://www.quicklogic.com/wp-content/uploads/2019/03/QL_S3AI_Platform_Data_Sheet_v1.1.pdf

```octave
P_compLow = Vdd * 17.52e-6;   % Compute core in standby mode
P_compHig = Vdd * 72 * 75e-6; % Compute core, assuming 72MHz and 75uA/MHz

f_comp = 1e-3;   % Hz; How often the compute core is activated by the RAMP detection trigger
T_comp = T_slow; % seconds; How long the compute core is on each time it is activated

P_compAvg = @(f_comp, T_comp) P_compHig * f_comp*T_comp ...  % power when active
                            + P_compLow * (1-f_comp*T_comp); % power when not active
```

* $P_\text{RAMP}$: RAMP average power

```octave
P_RAMP = Vdd * 10e-6;
```

* $P_{tot}$: Total system power

```octave
P_tot = @(f_rollAvg, f_press, T_press, f_audio, T_audio, f_comp, T_comp) ...
            P_rollAvg(f_rollAvg) ...
          + P_pressAvg(f_press, T_press) ...
          + P_audioAvg(f_audio, T_audio) ...
          + P_compAvg(f_comp, T_comp) ...
          + P_RAMP;
```

* $P_{base}$: Baseline power

```octave
P_base = P_tot(0,0,T_fast,0,T_fast,0,T_slow);
fprintf('Base power   = %duW\n',P_base*1e6);
fprintf('-------------------\n');
fprintf('  P_pressLow = %duW\n',P_pressLow*1e6);
fprintf('  P_audioLow = %duW\n',P_audioLow*1e6);
fprintf('  P_compLow  = %duW\n',P_compLow*1e6);
fprintf('  P_RAMP     = %duW\n',P_RAMP*1e6);
```
